```python

    #Python Optimization variables
batch_size = 258
mse_train = []
mse_test = []
epochs = 10
display_step = 1000
```

```python
n_movies = x_train.shape[1]

# Define a and b as placeholders
x = tf.placeholder(dtype=tf.float32,shape=[None, n_movies],name = "x-input")
y = tf.placeholder(dtype=tf.float32,shape=None,name = "y-input")
```

```python

    #Neurons

n_neurons_1 = 258
n_neurons_2 = 172
n_neurons_3 = 86
```

```python

    # Initializers
sigma = 1
weights = tf.contrib.layers.variance_scaling_initializer(mode='FAN_AVG', uniform=True,seed=None)
biases = tf.zeros_initializer()
```

```python
  
# Layer 1: Variables for hidden weights and biases
W_hidden_1 = tf.Variable(weights([n_movies, n_neurons_1]))
bias_hidden_1 = tf.Variable(biases([n_neurons_1]))

# Layer 2: Variables for hidden weights and biases
W_hidden_2 = tf.Variable(weights([n_neurons_1, n_neurons_2]))
bias_hidden_2 = tf.Variable(biases([n_neurons_2]))

# Layer 3: Variables for hidden weights and biases
W_hidden_3 = tf.Variable(weights([n_neurons_2, n_neurons_3]))
bias_hidden_3 = tf.Variable(biases([n_neurons_3]))

# Output layer: Variables for output weights and biases
W_out = tf.Variable(weights([n_neurons_3, 1]))
bias_out = tf.Variable(biases([1]))
```

```python
hidden_1 = tf.nn.relu(tf.add(tf.matmul(x, W_hidden_1), bias_hidden_1))
hidden_2 = tf.nn.relu(tf.add(tf.matmul(W_hidden_1, W_hidden_2), bias_hidden_2))
hidden_3 = tf.nn.relu(tf.add(tf.matmul(W_hidden_2, W_hidden_3), bias_hidden_3))


# Output layer (must be transposed)
out = tf.add(tf.matmul(hidden_3, W_out), bias_out)
```

```python
#print(W_hidden_3.graph) #check graphn name of variables
```

```python
# Cost function
mse = tf.reduce_mean(tf.squared_difference(out, y))
```

```python
# Optimizer
opt = tf.train.AdamOptimizer().minimize(mse)
```

```python
#predictions = multilayer_perceptron(x, weight_initializer, bias_initializer)
```

```python
# Setup plot
mp.ion()
fig = mp.figure()
ax1 = fig.add_subplot(111)
line1, = ax1.plot(y_test)
line2, = ax1.plot(y_test * 0.5)
mp.show()
```

```python
# Number of epochs and batch size



# Evaluation
#with tf.Session() as sess:
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())

for epoch in range(epochs):

    # Shuffle training data
    shuffle_indices = np.random.permutation(np.arange(len(y_train)))
    x_train = x_train[shuffle_indices]
    y_train = y_train[shuffle_indices]
```

```python
  # Minibatch training
for i in range(0, len(y_train) // batch_size):
    start = i * batch_size
    batch_x = x_train[start:start + batch_size]
    batch_y = y_train[start:start + batch_size]  
    
        # Run optimizer with batch
sess.run([opt], feed_dict={x: batch_x, y: batch_y})

    
```

```python
    # Show progress
    if np.mod(i, 50) == 0:
            # MSE train and test
            mse_train.append(sess.run(mse, feed_dict={x: x_train, y: y_train}))
            mse_test.append(sess.run(mse, feed_dict={x: x_test, y: y_test}))
            print('MSE Train: ', mse_train[-1])
            print('MSE Test: ', mse_test[-1])
            # Prediction
            pred = sess.run(out, feed_dict={x: x_test})
            line2.set_ydata(pred)
            mp.title('Epoch ' + str(e) + ', Batch ' + str(i))
            mp.pause(0.01)
            
# Print final MSE after Training
mse_final = sess.run(mse, feed_dict={x: x_test, y: y_test})
print(mse_final)
```
